```python
import pandas as pd
```

```python
 file = "../Resources/Cities_weather.csv"

weather_df = pd.read_csv(file)
weather_df.head()
```

```python
weather_df = weather_df.rename(columns = {"Unnamed: 0":"City_ID"})
weather_df = weather_df.dropna()
weather_df.head()
```

```python
html = weather_df.to_html(classes=["table-bordered", "table-striped", "table-hover"])
```

```python
text_file = open("../index.html", "w") 
text_file.write(html) 
text_file.close()
```
